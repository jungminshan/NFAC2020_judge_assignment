```python
import pandas as pd
import numpy as np
import sys
```

```python
#data = pd.read_excel("../2019_Conference/Identifers Registration for 15th Annual Conference.xlsx")
data = pd.read_excel("Identifers Registration for 15th Annual Conference.xlsx")

data = data.dropna(how='all')
```

```python
data.columns
```

```python
names = (data['Tracking Number'].to_list())
employee_types = (data['Employee Type'].to_list())
PIs = (data['Supervisor Name'].to_list())
laboratories = (data['Laboratory'].to_list())
buildings =  (data['Building'].to_list())
# rooms = (data['Room'].to_list())
# judges = (data['Judge'].to_list())
# presentation_types = (data['Presentation Type'].to_list())
research_keywords = (data['Research Keyword'].to_list())

judges = (data['Judge'].to_list()) #VS
presentation_types = (data['Presentation Type'].to_list()) #VS
```

```python
class Fellow():
    def __init__(self):
        self.name=''
        self.employee_type=''
        self.PI=''
        self.laboratory=''
#         self.building=''
#         self.room=''
#         self.presentation_type=''
        self.research_keyword=''
        self.judges=[]
        self.judging=[]
        
        self.judge='' #VS
        self.presentation_type='' #VS
```

```python
group_v = []
group_1 = []
group_1_nojudging = [] #VS
group_2 = []
group_2_nojudging = [] #VS

for i in range(len(names)):
    person = Fellow()
    person.name = names[i]
    person.employee_type = employee_types[i]
    person.laboratory = laboratories[i]
    person.PI = PIs[i]
    person.judge = judges[i] #VS
    person.presentation_type = presentation_types[i] #VS
    
    if person.employee_type == 'IRTA pre-doctoral' or person.employee_type == 'Post Baccalaureate Fellow':
        if person.judge == 'Y': #VS
            group_1.append(person) #VS
        else: #VS
            group_1_nojudging.append(person) #VS
#         group_1.sort(key=lambda x: x.employee_type)
    elif person.employee_type == 'Post Doctoral Fellow' or person.employee_type == 'Research Fellow' or person.employee_type == 'Clinical Fellow':
        if person.judge == 'Y': #VS
            group_2.append(person) #VS
        else: #VS
            group_2_nojudging.append(person) #VS
#         group_2.sort(key=lambda x: x.employee_type)
    else:
        group_v.append(person)
#         group_v.sort(key=lambda x: x.employee_type)
#     person.presentation_type=presentation_types[i]
```

```python
print('# of pre-doc and postbac: ', len(group_1))
print('# of postdoc, research, clinical fellows: ', len(group_2))
print('# of volunteer judges: ', len(group_v))

print("\nRules:")
print("- 1 pre-doc or postbac per postdoc, research, clinical abstract")
print("- pre-doc and postbac cannot judge each others\n")

group_v_maximum_judging = int(np.ceil((len(group_1) + len(group_2))/len(group_v)))
print('maximum # of abstracts assigned to volunteer judges: ', group_v_maximum_judging)
group_1_maximum_judging = int(np.ceil(len(group_2)/len(group_1)))
print('maximum # of abstracts assigned to pre-doc and postbac: ', group_1_maximum_judging)
group_2_maximum_judging = int(np.ceil(len(group_2)/len(group_1)))
print('maximum # of abstracts assigned to postdoc, research, clinical fellows: ', group_v_maximum_judging)

maximum_judges_per_abstract = 6 
```

```python
for person in group_1:
    if person.presentation_type == 'O' or person.presentation_type == 'P': #VS
        judges_from_group_v = [f for f in group_v if f.laboratory != person.laboratory and \
                              f.PI != person.PI and\
                              len(f.judging) <= group_v_maximum_judging]
        if len(judges_from_group_v) == 0:
            pass
        elif len(judges_from_group_v) <= 2:
            judges = np.random.choice(judges_from_group_v, len(judges_from_group_v), replace=False)
            person.judges.extend(judges)
            for judge in judges:
                judge.judging.append(person)
        elif len(judges_from_group_v) > 2:
            judges = np.random.choice(judges_from_group_v, 2, replace=False)
            person.judges.extend(judges)
            for judge in judges:
                judge.judging.append(person)
    
        total_judges = maximum_judges_per_abstract - len(person.judges)
        judges_from_group_2 = [f for f in group_2 if f.laboratory != person.laboratory and \
                              f.PI != person.PI and\
                              len(f.judging) < 6]
        if len(judges_from_group_2) == 0:
            pass
        else:
            judges = np.random.choice(judges_from_group_2, int(total_judges), replace=False)
            person.judges.extend(judges)
            for judge in judges:
                judge.judging.append(person)
    

for person in group_2:
    if person.presentation_type == 'O' or person.presentation_type == 'P': #VS
        judges_from_group_1 = [f for f in group_1 if f.laboratory != person.laboratory and\
                  f.PI != person.PI and\
                  len(f.judging) < int(np.ceil(len(group_2)/len(group_1)))]
        judge = np.random.choice(judges_from_group_1, 1)[0]
        person.judges.append(judge)
        judge.judging.append(person)
    
        judges_from_group_v = [f for f in group_v if f.laboratory != person.laboratory and \
                              f.PI != person.PI and\
                              len(f.judging) < 6]
        if len(judges_from_group_v) == 0:
            pass
        elif len(judges_from_group_v) <= 2:
            judges = np.random.choice(judges_from_group_v, len(judges_from_group_v), replace=False)
            person.judges.extend(judges)
            for judge in judges:
                judge.judging.append(person)
        elif len(judges_from_group_v) > 2:
            judges = np.random.choice(judges_from_group_v, 2, replace=False)
            person.judges.append(judges)
            for judge in judges:
                judge.judging.append(person)
    
        total_judges = maximum_judges_per_abstract - len(person.judges)
        judges_from_group_2 = [f for f in group_2 if f.laboratory != person.laboratory and\
                               f.PI != person.PI and\
                               len(f.judging) < 6]
        if len(judges_from_group_2) == 0:
            pass
        elif len(judges_from_group_2) < 5:
            judges = np.random.choice(judges_from_group_2, len(judges_from_group_2), replace=False)
            person.judges.extend(judges)
            for judge in judges:
                judge.judging.append(person)
        else:
            judges = np.random.choice(judges_from_group_2, total_judges, replace=False)
            person.judges.extend(judges)
            for judge in judges:
                judge.judging.append(person)
# #     print(person.judges)
```

```python
people = []
people.extend(group_1)
people.extend(group_2)
people.extend(group_v)

for person in group_1:
    print(len(person.judges), len(person.judging))
print('\n')
    
for person in group_2:
    print(len(person.judges), len(person.judging))
print('\n')
    
for person in group_v:
    print(len(person.judges), len(person.judging))
```
