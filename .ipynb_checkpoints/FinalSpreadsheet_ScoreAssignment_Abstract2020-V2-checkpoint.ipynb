```python
import pandas as pd
import numpy as np
import sys
```

```python
import os, shutil
list_rubrics = os.listdir('./NFAC2020_ReturnedRubrics_061820')
```

```python
dataframe_pavol = pd.read_excel('./061820_NFAC2020_JudgingRecords.xlsx')
disqualified = dataframe_pavol[dataframe_pavol["Eligible?" ] == "DISQUALIFIED"]
disqualified_list = disqualified['trackingNumber'].to_list()
print(disqualified_list)
```

```python
employee_type = dataframe_pavol.set_index('trackingNumber').to_dict()['Employee Type']
```

```python
from collections import defaultdict

submitted_judges = {}

presenter_scores = defaultdict(pd.DataFrame)

for rubric in list_rubrics:
    if rubric.endswith('.xlsx'):
        judge = rubric.split('_')[0]

        if judge == 'SC031452FE':
            judge = 'SC013452FE'
        submitted_judges[judge] = []

        df = pd.read_excel('./NFAC2020_ReturnedRubrics_061820/%s' % rubric, skiprows = 7, usecols='D, L:U', nrows = 12, index_col = 0)
        df = df.dropna(how='all', axis = 1).dropna(how='all', axis = 0)
        
        presenters = df.columns  
        new_presenters = []
        for presenter in list(presenters):
            presenter_orig = presenter
            if 'S' in presenter: ###
                presenter = presenter.lstrip(presenter[0: presenter.find('S')]) ###
            if presenter.startswith('\xa0'):
                presenter = presenter.strip('\xa0')
            if presenter.startswith('SCO'): ###
                presenter = presenter.replace('SCO', 'SC0')
            if not presenter.startswith('SC0'):
                presenter = presenter.replace('SC', 'SC0')
            if not presenter.startswith('SC01'):
                presenter = presenter.replace('SC0', 'SC01')
            if not presenter[-2:].isupper():
                presenter = presenter.replace(presenter[-2:], presenter[-2:].upper())
            presenter = presenter[0:10]
            submitted_judges[judge].append(presenter)
            new_presenters.append(presenter)
        
        df = df.rename(index = {np.nan: 'Total Score'})
        df.columns = new_presenters
        
        for presenter in new_presenters:
            presenter_scores[presenter] = pd.concat([presenter_scores[presenter], df[presenter]], axis = 1, sort = False)
        
```

```python
first_level_raw = df.index.to_list()
first_level = []
for i in first_level_raw:
    first_level.extend([i]*3)

second_level = ['Top3Mean', '4thScore', '5thScore']*len(df.index.to_list())

multilevel_indexes =  pd.MultiIndex.from_tuples(zip(first_level, second_level))
```

```python
scores_df = pd.DataFrame(index = multilevel_indexes)
```

```python
for item in presenter_scores.items():
    if item[0] not in disqualified_list and str(item[0]).startswith('SC0'):
        sorted_scores = item[1].sort_values(by='Total Score', axis=1,ascending=False)
        mean3top = sorted_scores.iloc[:, :3].mean(axis=1)
        score4 = sorted_scores.iloc[:, 3]
        try:
            score5 = sorted_scores.iloc[:, 4]
        except IndexError:
            score5 = np.nan
        idx = pd.IndexSlice
        scores_df.loc[idx[:, 'Top3Mean'], item[0]] = mean3top.values
        scores_df.loc[idx[:, '4thScore'], item[0]] = score4.values
        try:
            scores_df.loc[idx[:, '5thScore'], item[0]] = score5.values
        except AttributeError:
            scores_df.loc[idx[:, '5thScore'], item[0]] = score5
```

```python
scores_df = scores_df.sort_values(by = [('Total Score', 'Top3Mean'), ('Total Score','4thScore'), ('Total Score', '5thScore')], axis=1, ascending=[False, False, False])
```

```python
scores_df
```

```python
multilevel_indexes =  pd.MultiIndex.from_tuples([(' ', 'Employee Type')])
employee_types_list = [employee_type[column] for column in scores_df.columns.to_list()]
employee_types_df = pd.DataFrame(index=multilevel_indexes, columns = scores_df.columns.to_list())
employee_types_df.loc[idx[:, 'Employee Type'], :] = pd.DataFrame(data=employee_types_list).transpose().values
new_scores_df = pd.concat([employee_types_df, scores_df], axis = 0, sort = False)
```

```python
writer = pd.ExcelWriter('ScoresCategories_detailed.xlsx', engine='xlsxwriter')
new_scores_df.to_excel(writer, sheet_name = 'Scores', startcol=0)
writer.save()
```
