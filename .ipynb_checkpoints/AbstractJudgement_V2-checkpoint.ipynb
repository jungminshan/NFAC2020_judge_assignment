```python
import pandas as pd
import numpy as np
import sys
```

```python
data = pd.read_excel('Identifers Registration for 15th Annual Conference.xlsx')
data = data.dropna(how='all')
```

```python
class Person():
    def __init__(self):
        self.name=''
        self.employee_type=''
        self.supervisor_name=''
        self.laboratory=''
        self.building=''
        self.room=''
        self.judge=''
        self.presentation_type=''
        self.research_keyword=''

        self.judging=[]
        self.judges = []
```

```python
names = (data['Tracking Number'].to_list())
employee_types = (data['Employee Type'].to_list())
supervisor_names = (data['Supervisor Name'].to_list())
laboratories = (data['Laboratory'].to_list())
buildings =  (data['Building'].to_list())
rooms = (data['Room'].to_list())
judges = (data['Judge'].to_list())
presentation_types = (data['Presentation Type'].to_list())

people = []
postbac_list = []
postdoc_list = []
postbac_judges_list = []
postdoc_judges_list = []
postbac_presenters_list = []
postdoc_presenters_list = []
postbac_presenters_notjudging = [] #send email to remind they should judge if they want to partecipate to the award competition 
postdoc_presenters_notjudging = [] #send email to remind they should judge if they want to partecipate to the award competition
postbac_judging_notpresenting = [] #send email saying they cannot judge if they don't present (RULE from excel)
postdoc_judging_notpresenting = [] #send email asking if they are sure they want to judge and not present
other_judges_list = []
just_registered_list = []

presenters_list_to_judge=[]
postdoc_list_to_judge = []
postbac_list_to_judge = []

for i in range(len(names)):
    person = Person()
    people.append(person)
    person.name=names[i]      
    person.employee_type=employee_types[i]
    person.judge = judges[i]
    person.presentation_type = presentation_types[i]
    
    person.supervisor_name = supervisor_names[i]
    if str(person.supervisor_name) == 'nan':
        person.supervisor_name = 'other'
        
    person.laboratory = laboratories[i]
    
    if person.employee_type=='IRTA pre-doctoral' or person.employee_type=='Post Baccalaureate Fellow':
        postbac_list.append(person)
        if person.presentation_type == 'O' or person.presentation_type == 'P':
            postbac_presenters_list.append(person)
            if person.judge == 'Y':
                postbac_judges_list.append(person)
                presenters_list_to_judge.append(person)
                postbac_list_to_judge.append(person)
            else:
                postbac_presenters_notjudging.append(person)
        else:
            if person.judge == 'Y':
                postbac_judges_list.append(person)
                postbac_judging_notpresenting.append(person)
    elif 'Fellow' in person.employee_type:
        postdoc_list.append(person)
        if person.presentation_type == 'O' or person.presentation_type == 'P':
            postdoc_presenters_list.append(person)
            if person.judge == 'Y':
                postdoc_judges_list.append(person)
                presenters_list_to_judge.append(person)
                postdoc_list_to_judge.append(person)
            else:
                postdoc_presenters_notjudging.append(person)
        else:
            if person.judge == 'Y':
                postdoc_judges_list.append(person)
                postdoc_judging_notpresenting.append(person)
    else:
        if person.judge == 'Y':
            other_judges_list.append(person)
        else:
            just_registered_list.append(person)
 
judges_list_total = postbac_judges_list + postdoc_judges_list + other_judges_list
judges_list_nopostbac = postdoc_judges_list + other_judges_list
```

```python
print('Postbac judges: %s' % len(postbac_judges_list))
print('Postbac judging but not presenting: %s' % len(postbac_judging_notpresenting))
print('Postbac presenters: %s' % len(postbac_presenters_list))
print('Postbac presenters to be judged (because judging): %s' % len(postbac_list_to_judge))
print('Postbac presenters not judging: %s' % len(postbac_presenters_notjudging))
print('\n')

print('Postdoc judges: %s' % len(postdoc_judges_list))
print('Postdoc judging but not presenting: %s' % len(postdoc_judging_notpresenting))
print('Postdoc presenters: %s' % len(postdoc_presenters_list))
print('Postdoc presenters to be judged (because judging): %s' % len(postdoc_list_to_judge))
print('Postdoc presenters not judging: %s' % len(postdoc_presenters_notjudging))
print('\n')

print('Other judges: %s' % len(other_judges_list))
```

```python
import random

iteration1=1
while iteration1 <= 1000: 
    indexes = []
    n = random.randrange(0, len(postbac_judges_list), 1)
    for index in range(len(postbac_judges_list)):
        if index+n < len(postbac_judges_list):
            indexes.append(index+n)
        else:
            indexes.append(index+n-len(postbac_judges_list))
    for pbjudge in postbac_judges_list:
        pbjudge.judging = []
    for postdoc in postdoc_list_to_judge:
        postdoc.postbac_judges = [] 
    
    iteration2 = 1
    while len([pbjudge.judging for pbjudge in postbac_judges_list if len(pbjudge.judging)>1]) < len(postbac_judges_list) or len([postdoc.postbac_judges for postdoc in postdoc_list_to_judge if postdoc.postbac_judges!=[]]) < len(postdoc_list_to_judge):
        if iteration2 <= 1000:
            for postdoc in postdoc_list_to_judge:                                                                                                                               
                for index in indexes:
                    pbjudge = postbac_judges_list[index]
                    if postdoc.laboratory != pbjudge.laboratory and pbjudge not in postdoc.judges:
                        if [x for x in postbac_judges_list if len(x.judging) < len(pbjudge.judging)] != []: #to distribute equally over judges
                            pass
                        else:
                            pbjudge.judging.append(postdoc)
                            postdoc.postbac_judges.append(pbjudge)
                            break
        else:
            break
        iteration2+=1
        
    if len([pbjudge.judging for pbjudge in postbac_judges_list if len(pbjudge.judging)>1]) == len(postbac_judges_list) and len([postdoc.postbac_judges for postdoc in postdoc_list_to_judge if postdoc.postbac_judges!=[]]) == len(postdoc_list_to_judge):
        break
    else:
        if iteration1 == 1000:
            print("It was not possible to assign postbac judges respecting the rules")
        else:
            iteration1+=1
```

```python
import random

iteration1=1
while iteration1 <= 1000: 
    indexes = []
    n = random.randrange(0, len(judges_list_nopostbac), 1)
    for index in range(len(judges_list_nopostbac)):
        if index+n < len(judges_list_nopostbac):
            indexes.append(index+n)
        else:
            indexes.append(index+n-len(judges_list_nopostbac))
    for judge in judges_list_nopostbac:
        judge.judging = []
    for presenter in presenters_list_to_judge:
        presenter.judges = [] 
    
    iteration2 = 1
    while len([judge.judging for judge in judges_list_nopostbac if judge.judging != []]) < len(judges_list_nopostbac) or len([presenter.judges for presenter in presenters_list_to_judge if len(presenter.judges)>2]) < len(presenters_list_to_judge):
        if iteration2 <= 1000:
            for presenter in presenters_list_to_judge:
                for index in indexes:
                    judge = judges_list_nopostbac[index]
                    if presenter.laboratory != judge.laboratory and judge not in presenter.judges:
                        if [x for x in judges_list_nopostbac if len(x.judging) < len(judge.judging)] != []: #to distribute equally over judges
                            pass
                        else:
                            judge.judging.append(presenter)
                            presenter.judges.append(judge)
                            break
        else:
            break
        iteration2+=1

    if len([judge.judging for judge in judges_list_nopostbac if judge.judging != []]) == len(judges_list_nopostbac) and len([presenter.judges for presenter in presenters_list_to_judge if len(presenter.judges)>2]) == len(presenters_list_to_judge):
        break
    else:
        if iteration1 == 1000:
            print("It was not possible to assign judges respecting the rules")
        else:
            iteration1+=1
```

```python
postbac_list_to_judge_names = []
postbac_list_to_judge_laboratory = []
postbac_list_to_judge_employeetype = []
postbac_list_to_judge_judges_names = []
postbac_list_to_judge_judges_laboratory = []
postbac_list_to_judge_judges_employeetype = []
postbac_list_to_judge_judging_names = []
postbac_list_to_judge_judging_laboratory = []
postbac_list_to_judge_judging_employeetype = []

for presenter in postbac_list_to_judge:
    
    n_lines = max(len(presenter.judges), len(presenter.judging))
    
    postbac_list_to_judge_names.extend([presenter.name]+['']*(n_lines-1)+['\n'])
    postbac_list_to_judge_laboratory.extend([presenter.laboratory]+['']*(n_lines-1)+['\n'])
    postbac_list_to_judge_employeetype.extend([presenter.employee_type]+['']*(n_lines-1)+['\n'])
    
    n_lines_judges = n_lines - len(presenter.judges)
    n_lines_judging = n_lines - len(presenter.judging)

    postbac_list_to_judge_judges_names.extend([judge.name for judge in presenter.judges]+[' ']*n_lines_judges+['\n'])
    postbac_list_to_judge_judges_laboratory.extend([judge.laboratory for judge in presenter.judges]+[' ']*n_lines_judges+['\n'])
    postbac_list_to_judge_judges_employeetype.extend([judge.employee_type for judge in presenter.judges]+[' ']*n_lines_judges+['\n'])
    
    postbac_list_to_judge_judging_names.extend([j.name for j in presenter.judging]+[' ']*n_lines_judging+['\n'])    
    postbac_list_to_judge_judging_laboratory.extend([j.laboratory for j in presenter.judging]+[' ']*n_lines_judging+['\n'])
    postbac_list_to_judge_judging_employeetype.extend([j.employee_type for j in presenter.judging]+[' ']*n_lines_judging+['\n'])
```

```python
df = pd.DataFrame({'Postbac Presenter': postbac_list_to_judge_names,
                   'Laboratory': postbac_list_to_judge_laboratory,
                   'Employee type': postbac_list_to_judge_employeetype,
                   'Judges Identifier': postbac_list_to_judge_judges_names,
                   'Judges Laboratory': postbac_list_to_judge_judges_laboratory,
                   'Judges Employee type': postbac_list_to_judge_judges_employeetype,
                   'Judging Identifier': postbac_list_to_judge_judging_names,
                   'Judging Laboratory': postbac_list_to_judge_judging_laboratory,
                   'Judging Employee type': postbac_list_to_judge_judging_employeetype,
                  })

                  
writer = pd.ExcelWriter('presenters_spreadsheet.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Sheet1', index=False)
```

```python
postdoc_list_to_judge_names = []
postdoc_list_to_judge_laboratory = []
postdoc_list_to_judge_employeetype = []
postdoc_list_to_judge_judges_names = []
postdoc_list_to_judge_judges_laboratory = []
postdoc_list_to_judge_judges_employeetype = []
postdoc_list_to_judge_judging_names = []
postdoc_list_to_judge_judging_laboratory = []
postdoc_list_to_judge_judging_employeetype = []

for presenter in postdoc_list_to_judge:
    presenter.total_judges = presenter.judges + presenter.postbac_judges
    
    n_lines = max(len(presenter.total_judges), len(presenter.judging))
    
    postdoc_list_to_judge_names.extend([presenter.name]+['']*(n_lines-1)+['\n'])
    postdoc_list_to_judge_laboratory.extend([presenter.laboratory]+['']*(n_lines-1)+['\n'])
    postdoc_list_to_judge_employeetype.extend([presenter.employee_type]+['']*(n_lines-1)+['\n'])
    
    n_lines_judges = n_lines - len(presenter.total_judges)
    n_lines_judging = n_lines - len(presenter.judging)

    postdoc_list_to_judge_judges_names.extend([judge.name for judge in presenter.total_judges]+[' ']*n_lines_judges+['\n'])
    postdoc_list_to_judge_judges_laboratory.extend([judge.laboratory for judge in presenter.total_judges]+[' ']*n_lines_judges+['\n'])
    postdoc_list_to_judge_judges_employeetype.extend([judge.employee_type for judge in presenter.total_judges]+[' ']*n_lines_judges+['\n'])
    
    postdoc_list_to_judge_judging_names.extend([j.name for j in presenter.judging]+[' ']*n_lines_judging+['\n'])
    postdoc_list_to_judge_judging_laboratory.extend([j.laboratory for j in presenter.judging]+[' ']*n_lines_judging+['\n'])
    postdoc_list_to_judge_judging_employeetype.extend([j.employee_type for j in presenter.judging]+[' ']*n_lines_judging+['\n'])
```
